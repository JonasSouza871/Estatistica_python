## 4.2 Utilizando tabelas padronizadas para encontrar o valor de z:

```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
from scipy.stats import norm

# Criando um DataFrame vazio com os índices e colunas desejados
# O índice vai de 0.00 a 3.99 com incrementos de 0.01
# As colunas vão de 0.00 a 0.09 com incrementos de 0.01
tabela_normal_padronizada = pd.DataFrame(
    [],
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns=["{0:0.2f}".format(i / 100) for i in range(0, 10)])

# Preenchendo a tabela com os valores da distribuição normal acumulada (CDF)
for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        # Calcula o valor de Z somando o índice e a coluna
        Z = np.round(float(index) + float(column), 2)
        # Usa a função cdf (Cumulative Distribution Function) da scipy para obter a probabilidade
        # e a formata com 4 casas decimais
        tabela_normal_padronizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

# Renomeia o eixo das colunas para 'Z' para maior clareza
tabela_normal_padronizada.rename_axis('Z', axis = 'columns', inplace = True)

# Exibe a tabela resultante
tabela_normal_padronizada
```

### **Tabela Resumo – Distribuições Binomial, Poisson e Normal**

| Distribuição | Quando Usar (Condições) | Equação | Exemplo de Aplicação |
|--------------|-------------------------|---------|----------------------|
| **Binomial** | - Número fixo de tentativas \(n\) <br> - Dois resultados possíveis (sucesso ou fracasso) <br> - Probabilidade \(p\) constante <br> - Tentativas independentes | \( P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \) | Probabilidade de **3 caras** em 5 lançamentos de moeda |
| **Poisson** | - Eventos raros em intervalo fixo (tempo ou espaço) <br> - Taxa média \(\lambda\) constante <br> - Eventos independentes | \( P(X = k) = \dfrac{\lambda^k e^{-\lambda}}{k!} \) | Número de **chamadas** em 1 hora (média 10 chamadas/hora) |
| **Normal** | - Variável contínua <br> - Dados simétricos em torno da média \(\mu\) <br> - Grande número de observações (teorema central do limite) | \( f(x) = \dfrac{1}{\sigma \sqrt{2\pi}} e^{-\dfrac{(x-\mu)^2}{2\sigma^2}} \) | **Alturas** de pessoas adultas ou **erros de medição** |

---

### **Significado das Letras**

- **\(n\)** = número total de tentativas (binomial)  
- **\(k\)** = número de sucessos desejados  
- **\(p\)** = probabilidade de sucesso em cada tentativa (binomial)  
- **\(\lambda\)** = taxa média de ocorrência de eventos (Poisson)  
- **\(e\)** = constante de Euler (aprox. 2,718)  
- **\(\mu\)** = média da distribuição (normal)  
- **\(\sigma\)** = desvio padrão da distribuição (normal)  
- **\(x\)** = valor da variável aleatória (normal)

O valor do gasto médio dos clientes de uma loja de conveniência é de R$ 45,50. Assumindo que o desvio padrão dos gastos é igual a R$ 15,00, qual deve ser o tamanho da amostra para estimarmos a média populacional com um nível de significância de 10%?

Considere que o erro máximo aceitável seja de 10%.

```python
from scipy.stats import norm
u = 45.5
desvio_padrao = 15
e = u * 0.10
print(e)
significancia = 0.1
confianca = 1- (significancia) # a confiança sendo a significancia menos o total

Area_de_confianca = confianca + (significancia/2)
print(Area_de_confianca)

z = norm.ppf(Area_de_confianca)
print(z)

# A frormula do erro é a abaixo:
# e  = z * (desvio_padrao / np.sqrt(n))
# Isolando o n:

n = (z * (desvio_padrao / e))**2
print(n) # Numero de amostras
```
