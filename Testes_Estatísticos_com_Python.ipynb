## 3.1. O efeito do tamanho de uma amostra
------------

```python
notas_filmes_mais_de_10_votos = media_por_filme_filtrada_mais_de_10.values
notas_filmes_mais_de_10_votos.mean()
```

- Vemos que a nota a nota media dos filmes é 3.44

- Porem esses valores podem ser alterados de acordo com a amostra. Por exemplo em uma casa com 4 pessoas todo mundo pode ter dado uma nota 1, onde cada amostra é dependente entre si.

- O teste independente consiste em pegar varias pessoas, que viram em locais diferentes, em momentos diferentes e calcular a probabilidade de cada um da notas parecidas

- Porem nunca se pode generalizar com poucos dados, tem que ter uma quantidade razoavel para chegar numa conclusão (Igualmente feito no dataset)

```python
# Calculando a media das primeiras 3 amostras:
notas_filmes_mais_de_10_votos[0:2].mean()
```

```python
# Calculando a media das 15 primeiras amostras
notas_filmes_mais_de_10_votos[0:14].mean()
```

```python
#total de notas
len(notas_filmes_mais_de_10_votos)
medias_amostras = list()
# faz um for para ver todas as medias de todas as amostras possiveis
for i in range(1, len(notas_filmes_mais_de_10_votos)):
    medias_amostras.append(notas_filmes_mais_de_10_votos[0:i].mean()) # e abaixo temos a media de todas as amostras
    # todas as medias de cada amostra armazenado dentro da list amarzenas amostras

notas_filmes_mais_de_10_votos
```

- Plotando essas distribuição de media de amostras:

```python
plt.plot(medias_amostras) #plota o grafico com o comportamento das media
```

- vemos que a media começa bem instave e vai se estabilizando conforme uma maior quantidade de amostras

- logo uma amostra pequena ou mal escolhida, não é relevante para o estudo no geral, pois não mostra a media real.

- Visualizar se a amostra representa o real

- Comportamento muito parecido com uma distribuição normal

## 3.2. Intervalo de confiança da média com t e zteste

----------------------

- Qual intervalo de confiança para a media dos filmes? A partir de quanto é uma media representativa?

```python
from statsmodels.stats.weightstats import zconfint #teste de confiança para z
# FAZER ISSO AQUI NA MAO DEPOIS, SALVAR
zconfint(notas_filmes_mais_de_10_votos)
```

- Intervalo de confiança entre 3.42 e 3.46

- Garante que médias de amostras nesse intervalo tendem a representar a realidade

- Se a amostras fossem pequenas usariamos o t teste conforme estudado no curso de estatistica anterior, como é uma quantidade maior usamos o z_teste e o t teste tende a não da um bom resultado.

```python
# testando t_teste:
from statsmodels.stats.weightstats import DescrStatsW

test = DescrStatsW(notas_filmes_mais_de_10_votos) #cria o objeeto para realizar o teste
test.tconfint_mean() #t teste para a media
```

-  Vemos que apesar de não ser muito indicado o t teste apresentou um resultado muito parecido com o z teste para esse caso.

- 3.92 de media. Porem não podemos ainda confiar nesses valores por causa justamente das amostras , é preciso verificar via intervalos:

```python
zconfint(notas1)
```

-  Intervalo de confiança entre 3.8 e 4.03 grantindo 95% de confiança

- vamos fazer um teste estatistico agora para conferir se a media do filme de toy story é igual a média dos filmes em geral

- Para isso vamos fazer um z teste, que vai retonar o valor de z_teste estatistico e tambem o value.

- Lembrando que temos que rejeitar a hipotese se P_value <= Significancia

```python
from statsmodels.stats.weightstats import ztest
z_teste,p_valor = ztest(notas1, value=3.43) # media dos filmes igual a 3.43
  #primeiro termo de saida será o z estatistico e o segundo o p valor
print(z_teste)
print(p_valor)
```

```python
# teste da hipotese h0:
significancia = 0.05 #significancia da propria biblioteca (5%)
p_valor <= significancia
```

- True, logo temos que rejeita H0, e recusar a ideia que a media do filme de toy story se comporta igual a media dos filmes em geral.

- Aceitamos H1 de hipotese alternativa, que dirá que a media de notas do filme de toy story não segue a media geral dos outros filmes.

- Esse teste não é tão ideal visto que a media é um numero muito especifico, problema clássico de  big data (grande quantidade de dados) para resolver isso é necessario fazer outros testes, como menor que e maior que (Unicaudal )

## 4.2. Problemas de uma amostra muito pequena

------------

- Fazendo a media de amostras para um unico filme

```python
#total de notas
len(notas1)
medias_amostras_notas_toy_story = list()
# faz um for para ver todas as medias de todas as amostras possiveis
for i in range(1, len(notas1)):
    medias_amostras_notas_toy_story.append(notas1[0:i].mean()) # e abaixo temos a media de todas as amostras
    # todas as medias de cada amostra armazenado dentro da list amarzenas amostras

medias_amostras_notas_toy_story
```

```python
plt.plot(medias_amostras_notas_toy_story) #visualizar o comportamento das medias
```

- fazendo o teste estatistico para media de cada amostra:

```python
#total de notas
len(notas1)
# faz um for para ver todas as medias de todas as amostras possiveis
def calcular_teste(i):
     # Calcula a média da amostra até o índice i
     media = (notas1[0:i].mean())
     # Realiza o z-teste usando a amostra (notas1[0:i]) e o valor a ser comparado (3.43)
     z_teste,p_valor = ztest(notas1[0:i], value=3.43) # media dos filmes igual a 3.43
     return(media, z_teste,p_valor) #vai retornar 3 valors, a media obtida
     #o z_Teste estatistico calculado para cada valor
     # e o p vaor para cada media

medias_amostras_notas_toy_story_teste_z = [calcular_teste(i) for i in range(1, len(notas1))]
medias_amostras_notas_toy_story_teste_z # 3 saidas
```

- Para mostrar na saida o valor qual amostra é tambem

```python
#total de notas
len(notas1)
# faz um for para ver todas as medias de todas as amostras possiveis
def calcular_teste(i):
     # Calcula a média da amostra até o índice i
     media = (notas1[0:i].mean())
     # Realiza o z-teste usando a amostra (notas1[0:i]) e o valor a ser comparado (3.43)
     z_teste,p_valor = ztest(notas1[0:i], value=3.43) # media dos filmes igual a 3.43
     return(i, media, z_teste,p_valor) #vai retornar 4 valors, a media obtida
     #o z_Teste estatistico calculado para cada valor
     # e o p vaor para cada media
     # a amostra

medias_amostras_notas_toy_story_teste_z = [calcular_teste(i) for i in range(1, len(notas1))]
medias_amostras_notas_toy_story_teste_z # 3 saidas
```

```python
import numpy as np

valores= np.array(medias_amostras_notas_toy_story_teste_z) #convertendo a lista em um array numpy
plt.plot(valores[:,0],valores[:,1]) #plota o grafico com o comportamento das media
plt.plot(valores[:,0],valores[:,3]) #comportamento do p valor
```

- no grafico acima vemos a comparação entre os valores das medias e o p valor


- O p valor deve ser menor que a significancia (0.05) para rejeitar a hipotese H0 (a media de notas do toy story é igual a media dos filmes em gerais (3.42))

```python
valores= np.array(medias_amostras_notas_toy_story_teste_z) #convertendo a lista em um array numpy
plt.plot(valores[:,0],valores[:,3]) #comportamento do p valor
# adicionando a linha de significicancia:

significancia = 0.05 #5%

plt.hlines(y=significancia, xmin=0, xmax=len(notas1), color="black") #adiciona a linah de significancia
```

-  Então conferimos que em alguns momentos o HO é aceito, logo a quantidade de amostras irá influencias muito nas hipoteses.

- Se a gente usasse um conjunto pequeno (menor que 50) provavelmente iamos concluir que a media do filme do toystory é igual que a media geral dos filmes

- Por isso a necessidade de ter mais dados, mais informações, mais conjuntos de amostras para ter um resultado melhor.

- PARA FAZER O TESTE Z PARA DUAS AMOSTRAS:

- O z-teste para duas amostras é importante para verificar se há diferença estatisticamente significativa entre as médias de dois grupos independentes.

```python
from scipy.stats import ttest_ind # importa a função responsávle por fazer o teste de duas amostras
```

```python
ttest_ind(notas_filmes_mais_de_10_votos, notas1)
```

Esse resultado indica que a estatística t é -12,42 e o valor-p é extremamente pequeno (≈ 2,4 × 10⁻³⁴), o que significa que há evidência fortíssima de que as médias das duas amostras são diferentes.

As media são extremamentes diferentes e o vlaor é menor que a significancia (0.05) o que leva a crer que as medias não sao iguais
