```python
import seaborn as sns
import matplotlib.pyplot as plt

boxplot = sns.boxplot(x=tmdb['vote_average'])
boxplot.set(xlabel = "Nota média")
```

- Logo,algo que chama a atenção é ter filmes com media 0 e 10, o que seria impossivel, visto que todas as pessoas que votaram teriam que dar 0 ou 10 para que isso acontecesse.

```python
tmdb.query("vote_average == 0")
```

-  Filtando os dados para valer apenas os que tem mais de 10 votos.

```python
tmdb_com_mais_de_10_votos = tmdb.query("vote_count > 10") #filtrando para mais de 10
tmdb_com_mais_de_10_votos.head(10)
```

```python
ax = sns.histplot(tmdb_com_mais_de_10_votos['vote_average'], stat= 'count')
ax.set(xlabel = "Nota Média", ylabel = "Quantidade de Votos")
ax.set_title("Nota Média dos filmes com mais de 10 votos")
```

## 1.2. visualizando e limpando o movielens

-----

```python
notas = pd.read_csv("ratings.csv")
notas.head()
notas.query("movieId == 1") # aparece somente as notas de cada usuario para o filme com ID = 1
```

```python
media_por_filme = notas.groupby("movieId").mean()["rating"]
media_por_filme
```

```python
ax = sns.distplot(media_por_filme.values)
ax.set(xlabel = "Nota Média", ylabel = "Densidade")
ax.set_title('Média de votos no  Movielens')
```

- vamos que tem alguns filmes com nota 5 o que pode indicar quantidade de avaliaçoes baixas.
- Para testar isso:

```python
quantidade_de_notas_por_filme = notas.groupby("movieId").count()
quantidade_de_notas_por_filme.head()
```

```python
filmes_filtrados_quantidade_maior_que_10 = quantidade_de_notas_por_filme.query("rating > 10").index #filtrando apenas os index com mais de 10
filmes_filtrados_quantidade_maior_que_10.values #aparece todos os index( ID) com mais de 10 votos
```

```python
media_por_filme_filtrada_mais_de_10 = media_por_filme.loc[filmes_filtrados_quantidade_maior_que_10.values] #localiza os ids filtrados acima
media_por_filme_filtrada_mais_de_10
```

- vemos que de 9724 items no sem filtrar caiu para 2121 no filtrado, diminuindo bem a quantidade de IDs de filmes

```python
ax = sns.distplot(media_por_filme_filtrada_mais_de_10.values)
ax.set(xlabel = "Nota Média", ylabel = "Densidade")
ax.set_title('Média de votos no  Movielens')
```

- Comparando o antes e depois da filtragem dos IDs via subplot do matplot lib

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1) # linha no subplot, coluna no subplot e posição na grade
ax1 = sns.histplot(media_por_filme.values, stat='count')
ax1.set(xlabel="Nota Média", ylabel="Quantidade de Filmes", title="Média de votos no Movielens (Sem Filtro)")

plt.subplot(1, 2, 2) # como queremos o subplot ao lado iremos mudar somente a posição na grade
ax2 = sns.histplot(media_por_filme_filtrada_mais_de_10.values, stat='count')
ax2.set(xlabel="Nota Média", ylabel="Quantidade de Filmes", title="Média de votos no Movielens (Com Filtro > 10 votos)")
```
