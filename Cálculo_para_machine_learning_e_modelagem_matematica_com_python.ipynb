## 1.4. Expressões algébricas em python

- Uso da biblioteca sympy

A SymPy é uma biblioteca de código aberto em Python desenvolvida para realizar cálculos de matemática simbólica. Diferente de ferramentas numéricas, ela permite manipular expressões matemáticas de forma algébrica, oferecendo recursos avançados para estudantes, engenheiros e pesquisadores. Entre suas funcionalidades estão a resolução de equações algébricas e diferenciais, simplificação de expressões e cálculos de derivadas, integrais, limites e séries.

```python
import sympy as sy
```

```python
x = sy.Symbol("x") #declara variavel x com o simbolo x
```

sy.Symbol("x") cria um símbolo matemático chamado "x" que pode ser manipulado simbolicamente (derivado, integrado, substituído etc.) dentro das expressões do SymPy.

- Como se fosse feito a mão e esse x fosse um valor que pode ser somado, exemplo:

```python
x + x # x mais x igual a 2x. sem a biblitoeca não é possivel fazer esse algebrismo
```

```python
(x**3 + 2*x**2 + 3*x) + (4*x**3 - 3*x**2)
```

-  Declarar outras variáveis :

```python
y = sy.Symbol("y")
z = sy.Symbol("z")
```

```python
(x + 2*y) * (5 - 2*z)
```

- vemos que ele não fez o calculo (expansão do valor) e sim deixou da forma fatorada conforme foi entre. para alterar isso:

```python
exemplo = (x + 2*y) * (5 - 2*z)
```

```python
expansao = sy.expand(exemplo)
expansao
```

- agora sim com a multiplicação realizada.

- Fazendo o produto da soma pela diferença :

```python
exemplo2 = (x+y)*(x-y)
exemplo2 = sy.expand(exemplo2)
exemplo2
```

- Quadrado da soma:

```python
exemplo3 = (x+y)**2
exemplo3 = sy.expand(exemplo3)
exemplo3
```

-  Para fazer fatoração:

```python
exemplo5 = x**2 + 2*x*y + y**2
exemplo5 = sy.factor(exemplo5) #faz a fatoração do valor
exemplo5
```

```python
exemplo6 = x**3 + 3*x**2*y + 3*x*y**2 + y**3
exemplo6 = sy.factor(exemplo6)
exemplo6
```

-  Divisão de polinomios:

## 1.10. Funções logarítmicas

-----

image.png

```python
sy.plot(sy.log(x,1/4), (x,0.1,6)) #log de x na base 1/4 , com x variando entre 0.1 e 6 , ou seja:
# sy.log(x, 1/4) significa logaritmo de x na base 1/4.
# Lembre: log_b(x) responde à pergunta: "qual expoente y faz b^y = x?"
# Então, log_(1/4)(x) é o número y tal que (1/4)^y = x.
#
# Exemplo: se x = 0.25
# Queremos resolver (1/4)^y = 0.25.
# Mas 0.25 = 1/4, então y = 1.
# Logo, log_(1/4)(0.25) = 1.
```

- exemplo função crescente, considerando x > 1

```python
sy.plot(sy.log(x,3), (x, 0.1,10))
# sy.log(x, 3) significa logaritmo de x na base 3.
# Em outras palavras: log_3(x) é o expoente y que satisfaz 3^y = x.
#
# Características:
# - Como a base 3 é maior que 1, a função log_3(x) é CRESCENTE.
# - log_3(1) = 0, pois 3^0 = 1.
# - log_3(3) = 1, pois 3^1 = 3.
# - log_3(9) = 2, pois 3^2 = 9.
#
# Exemplo: se x = 9, temos log_3(9) = 2, já que 3^2 = 9.
```

- Plotando uma crescente e decrescente:

```python
sy.plot(sy.log(x,2), sy.log(x,1/2), (x,0.1,10)) # x na base 2 e x na base meio variando de 0 a 10.
```

## 1.13. Funções racionais
------

- Função expressa por razões de polinômios

- Dividida por assintonas verticais e horizontais.

image.png

- Se n < m, a assintona horizontal será o próprio eixo x
- se n > m, a assintona horizintal será inexistente
- Se n = m, terá assintona vertical e horizontal.

- exemplo onde n < m e o eixo x é a assintona horizontal:

```python
funcao_racional = 1/x
sy.plot(funcao_racional, (x, -1.5, 1.5))
```

- exemplo onde m = n , e tem ambas assintonas.

```python
funcao_racional = (4-2*x)/(1-x)
sy.plot(funcao_racional, (x, -2,2))
```
