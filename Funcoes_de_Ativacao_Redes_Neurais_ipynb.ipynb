# 1. Função Sigmóide

-------

 - Transforma função linear em não linear delimitando entre 0 e 1.

- Adiciona não lineariedade em modelos, permitindo o aprendizado.

image.png

- Classifcição em duas possíveis classes

- Valores após a passagem sempre serão positivos.

```python
import sympy as sy
x = sy.Symbol('x')
from sympy import plotting
import math
```

```python
funcao_sigmoide = 1/ (1+ math.e**(-x))
funcao_sigmoide
```

```python
sy.plot(funcao_sigmoide, (x, -10, 10))
```

- Acima de 0,5 classifica como 1 e abaixo de 0,5 se classifica como 0. Deixando apenas duas opçoes de classificação

**Exemplo : transformando uma linear em não linear**

```python
funcao_linear = 2*x + 0.3
funcao_sigmoide = 1/ (1+ math.e**(-funcao_linear))
funcao_sigmoide
```

```python
p1 = sy.plot(funcao_sigmoide, (x, -10, 10), title='Função Sigmoide', show=False)
p2 = sy.plot(funcao_linear, (x, -10, 10), title='Função Linear', show=False)
p1.show()
p2.show()
```

- Para plotar lado a lado:

```python
import matplotlib.pyplot as plt
import numpy as np

valores_x = np.linspace(-10, 10, 400) #de -10 a 10 com 4300 pontos entre
fig, axes = plt.subplots(1, 2, # 1 linha, 2 colunas
                         figsize=(10, 4))  #tamanho da imagme

funcao_linear = 2*(valores_x) + 0.3
funcao_sigmoide = 1/ (1+ math.e**(-funcao_linear))

axes[0].plot(valores_x, funcao_linear)
axes[0].set_title("Função Linear")

axes[1].plot(valores_x, funcao_sigmoide)
axes[1].set_title("Função Sigmoide")

plt.show()
```

# 2. Função Softmax

--------------

- Não classifica somente binárias

- Transforma cada saída para cada classe para valores entre 0 e 1

- Também divide pela soma das saídas

content_66c134e9-a34b-44d9-aa63-85dcf83f516e-3.png

```python
softmax_x = sy.E**(x) / (sy.E**(2*x+1) + sy.E**(x) + sy.E**(3*x))
softmax_x
```

```python
sy.plot(softmax_x, (x,-10,10))
```

mostra que a função softmax tem comportamento sigmoidal, tendendo a 0 para valores muito negativos e a 1 para valores muito positivos, com transição suave no centro.

**Exercícios 1 :**

Um modelo de machine learning simples precisa classificar uma imagem entre três
categorias: "Cachorro", "Gato" e "Pássaro". A camada de saída do modelo (antes da ativação)
produziu os seguintes scores (logits):

- Cachorro: 2.0
- Gato: 1.0
- Pássaro: 0.1

**a) Calcule a probabilidade de cada classe usando a função softmax:**

usando o for

```python
Cachorro = 2
Gato = 1
Passaro = 0.1
Scores_animais = [Cachorro, Gato, Passaro]

for x in Scores_animais:
  softmax = (math.e**x)/(math.e**(Cachorro) + math.e**(Gato) + math.e**(Passaro))
  print(softmax * 100)
```

passo a passo:

```python
soma_denominador = math.e**(Cachorro) + math.e**(Gato) + math.e**(Passaro) #soma de todas as classes

classe_cachorro = ((math.e**Cachorro)/soma_denominador ) *100 #Classe especifica de cachorro dividido pelo total de classe
classe_gato = ((math.e**Gato)/soma_denominador) *100 # Classe especifica de gato dividido pelo total de classe
classe_passaro = ((math.e**Passaro)/soma_denominador) * 100 #Classe especifica de passaro peelo total de classe

print("Probabilidade Cachorro :",classe_cachorro, "%") #probabilidade Cachorro perante o total
print("Probabilidade Gato :",classe_gato, "%") # Probabilidade gato perante o total
print("Probabilidade Passaro:",classe_passaro, "%") #Classe passaro perante o total
```

**b) Interprete os resultados. Qual classe o modelo previu como a mais provável e com qual
confiança?**


- Pelos resultados chegamos a conclusão que a imagem tem maior probabilidade de se tratar de um cachorro (65,9%) enquanto os outros animais apresentam probabilidades menores.

Aplicação em Processamento de Linguagem Natural NLP - Predição

**Cenário: Um modelo de linguagem simples está tentando prever a próxima palavra na frase "O
céu é...". Ele tem um vocabulário limitado de quatro palavras possíveis: "azul", "verde",
"vermelho", "alto". Os scores (logits) gerados pelo modelo para a próxima palavra são:**

- azul: 3.2
- verde: 1.5
- vermelho: 0.5
- alto: 2.8


**a) Utilize a função softmax para converter esses scores em um vetor de probabilidades.**

```python
azul = 3.2
verde = 1.5
vermelho = 0.5
alto = 2.8
scores_palavras = [azul, verde, vermelho, alto]

for x in scores_palavras:
  softmax = (math.e**x)/(math.e**(azul) + math.e**(verde) + math.e**(vermelho) + math.e**(alto))
  print(softmax * 100)
```

```python
soma_denominador = math.e**(azul) + math.e**(verde) + math.e**(vermelho) + math.e**(alto) #soma de todas as classes

classe_azul = ((math.e**azul)/soma_denominador)*100 #Classe especifica de azul dividido pelo total das classes
classe_verde = ((math.e**verde)/soma_denominador)*100 # Classe especifica de verde dividido pelo total das classes
classe_vermelho = ((math.e**vermelho)/soma_denominador)*100 #Classe especifica de vermelho peelo total das classes
classe_alto = ((math.e**alto)/soma_denominador)*100 #Classe especifica de alto peelo total das classes

print("Probabilidade azul :",classe_azul, "%") #probabilidade azul perante o total
print("Probabilidade verde :",classe_verde, "%") # Probabilidade verde perante o total
print("Probabilidade vermelho:",classe_vermelho, "%") #Probabilidade vermelho perante o total
print("Probabilidade alto:",classe_alto, "%") #Probabilidade alto perante o total
```

**b) Qual palavra é a mais provável de ser a próxima?**

É mais provável que a próxima palavra seja a palavra "azul" com 52% de probabilidade de ser ela.

**c) Se o modelo estivesse usando uma função de ativação como a sigmoide (aplicada a cada
saída individualmente), por que ela não seria a escolha ideal para esta tarefa?**

Não poderia ser sigmoide por conta que na sigmoide é ultizado para casos binários onde o resultado tende a ser 1 ou 0 (sim ou não de acordo com cada caso), no caso do nosso projeto temos várias classes.

A softmax considera a relação entre todas as classes simultaneamente, não apenas cada
uma isoladamente.

Se fizesse isolodamente cada termo via sigmoide nada garantiria que a soma total seria 1 ( ou 100%)


```python
#aplicando sigmoide
scores_palavras = [azul, verde, vermelho, alto]
for x in scores_palavras:
  funcao_sigmoide = ( 1/ (1+(math.e)**(-(x))) )*100
  print("Probabilidade de ser aprovado em", funcao_sigmoide, "%")
```

- Como vemos, os valores não saem padronizado, não sendo possível definir o mais provavel em 1 caso.

**Exercício 1**


Cenário: Uma empresa desenvolveu um sistema de análise de sentimentos que usa a função
tanh para classificar comentários. Diferente da sigmoide, a tanh produz valores negativos para
sentimentos negativos e positivos para sentimentos positivos, com zero representando
neutralidade.


- Comentário A "Produto excelente!" → score = 1.2
- Comentário B "Não gostei muito" → score = -0.7
- Comentário C "Produto OK" → score = 0.0
- Comentário D "Fantástico!" → score = 2.3
- Comentário E "Péssimo produto" → score = -1.5

**a) Calcule a probabilidade para cada score usando a definição exponencial.**

```python
comentario_1 = 1.2
comentario_2 = -0.7
comentario_3 = 0.0
comentario_4 = 2.3
comentario_5 = -1.5

valores_comentarios = [comentario_1, comentario_2, comentario_3, comentario_4, comentario_5] #vetor com todos os comentarios
vetor_tanh = []

for i,x in enumerate(valores_comentarios): #para ordenar os comentarios
  tanh = (2 / (1 + sy.E**(-2*x))) - 1
  vetor_tanh.append(tanh)
  print(f"comentario {i+1}",tanh)
print(vetor_tanh)
```

**b) Converta os valores tanh para probabilidades usando normalização**

```python
for i,x in enumerate(vetor_tanh):
  probabilidade = (x + 1)/2
  print(f"comentario {i+1}",probabilidade*100, "%")
```

**c) Classifique cada comentário baseado no sinal da saída tanh**

```python
for i,x in enumerate(vetor_tanh):
  if x > 0:
    print(f"comentario {i+1} positivo")
  elif x == 0:
    print(f"comentario {i+1} neutro")
  else:
    print(f"comentario {i+1} negativo")
```
